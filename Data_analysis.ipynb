Data Preprocesing Techniques

```python
array = df1.values                               #scaling the data... setting all values in the range  of (0,1)
from numpy import set_printoptions               
from sklearn import preprocessing
data_scaler = preprocessing.MinMaxScaler(feature_range = (0,1))
data_rescaled = data_scaler.fit_transform(array)
```

```python
set_printoptions(precision=1)
print(data_rescaled)
```

```python
data_normalizer = preprocessing.Normalizer(norm = 'l1').fit_transform(array)     #l1 normalisation where row values sumup = 1
print(data_normalizer)
```

```python
data_normalizer = preprocessing.Normalizer(norm = 'l2').fit_transform(array)     #l2 normalisation where row values induvidual square sumup = 1
print(data_normalizer)
```

```python
data_binarizer = preprocessing.Binarizer(threshold=0.5).fit_transform(array)  #Binarization where the values are rounding up to the closest vlaue 0 or 1
print(data_binarizer)
```

```python
data_scaler1 = preprocessing.StandardScaler().fit(array)
data_rescaled1 = data_scaler1.transform(array)                        #standarization based on mean and standard deviation
print(data_rescaled1)
```

```python
#label Encoding

input_names = ['red','blue','green','orange','black','white','yellow']
encoder = preprocessing.LabelEncoder()
labelencoder1 = encoder.fit(input_names)

test_labels = ['green','red','black']
encoded_values = encoder.fit(test_labels)
print("\nLabels =", test_labels)

encoded_values = [3,0,4,1]
print("\nEncoded values =", encoded_values)
```
